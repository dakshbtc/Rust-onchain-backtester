```python
import pandas as pd
import numpy as np
from numba import njit

# Assume 'df' is a pre-existing DataFrame loaded with your data
# with the necessary columns: 'Date', 'Token Address', 'Token Price', 
# 'SOL in Curve', and 'Transaction Type'.

# --- Numba functions ---
@njit
def pct_change(arr):
    result = np.empty_like(arr, dtype=np.float64)
    result[0] = np.nan
    for i in range(1, len(arr)):
        if arr[i-1] != 0:
            result[i] = (arr[i] - arr[i-1]) / arr[i-1]
        else:
            result[i] = np.nan
    return result

@njit
def delta_array(arr):
    result = np.empty_like(arr, dtype=np.float64)
    result[0] = np.nan
    for i in range(1, len(arr)):
        result[i] = arr[i] - arr[i-1]
    return result

@njit
def rolling_std(arr, window):
    n = len(arr)
    result = np.full(n, np.nan)
    for i in range(window-1, n):
        window_data = arr[i-window+1:i+1]
        # Skip window if it contains NaN
        if np.any(np.isnan(window_data)):
            continue
        mean = np.mean(window_data)
        # Ensure we don't divide by zero if window size is 1
        denom = window - 1 if window > 1 else 1
        if denom == 0: continue
        result[i] = np.sqrt(np.sum((window_data - mean)**2) / denom)
    return result

# --- Compute stats per token ---
def compute_stats(df_token, window=24):
    df_token = df_token.sort_values('Date').copy()
    
    # Ensure there's enough data for a rolling window
    if len(df_token) < window:
        # Not enough data, return empty dataframes to avoid errors
        return pd.DataFrame(), pd.DataFrame(), pd.DataFrame()

    price_arr = df_token['Token Price'].values
    sol_arr = df_token['SOL in Curve'].values
    
    # Numba computations
    df_token['Token_Returns'] = pct_change(price_arr)
    df_token['Volatility'] = rolling_std(df_token['Token_Returns'].values, window) * np.sqrt(window)
    df_token['Delta_SOL'] = delta_array(sol_arr)
    
    # Volatility bucket
    # Using .dropna() before qcut to avoid errors with NaNs from rolling window
    vol_series = df_token['Volatility'].dropna()
    if vol_series.nunique() > 2:
        try:
            df_token['Vol_Bucket'] = pd.qcut(
                vol_series, q=3,
                labels=['LowVol','MidVol','HighVol'],
                duplicates='drop'
            )
        except ValueError: # Catch qcut errors explicitly
            df_token['Vol_Bucket'] = "NA"
    else:
        df_token['Vol_Bucket'] = "NA"
    
    # Flow bucket
    flow_series = df_token['Delta_SOL'].dropna()
    if flow_series.nunique() > 2:
        try:
            df_token['Flow_Bucket'] = pd.qcut(
                flow_series, q=3,
                labels=['Outflow','Neutral','Inflow'],
                duplicates='drop'
            )
        except ValueError:
             df_token['Flow_Bucket'] = "NA"
    else:
        df_token['Flow_Bucket'] = "NA"
        
    # Fill NA buckets if qcut created any from dropna
    if pd.api.types.is_categorical_dtype(df_token['Vol_Bucket']):
        df_token['Vol_Bucket'] = df_token['Vol_Bucket'].cat.add_categories('NA').fillna('NA')
    else:
        df_token['Vol_Bucket'] = df_token['Vol_Bucket'].fillna('NA')

    if pd.api.types.is_categorical_dtype(df_token['Flow_Bucket']):
        df_token['Flow_Bucket'] = df_token['Flow_Bucket'].cat.add_categories('NA').fillna('NA')
    else:
        df_token['Flow_Bucket'] = df_token['Flow_Bucket'].fillna('NA')
    
    # Aggregations
    vol_stats = df_token.groupby('Vol_Bucket')['SOL in Curve'].agg(['mean','median','std','min','max'])
    cross_stats = df_token.groupby(['Vol_Bucket','Transaction Type'])['SOL in Curve'].agg(['mean','count'])
    flow_vol_stats = df_token.groupby(['Vol_Bucket','Flow_Bucket'])['Delta_SOL'].agg(['mean','count','std'])
    
    return vol_stats, cross_stats, flow_vol_stats

# --- Loop through all tokens ---
results = []
for token, df_token in df.groupby("Token Address"):
    try:
        vol_stats, cross_stats, flow_vol_stats = compute_stats(df_token)
        # Only append if the stats are not empty
        if not vol_stats.empty:
            results.append({
                "Token": token,
                "VolStats": vol_stats,
                "CrossStats": cross_stats,
                "FlowVolStats": flow_vol_stats
            })
    except Exception as e:
        print(f"Skipping token {token} due to error: {e}")

# --- Combine results ---
# Check if results list is not empty before concatenating
if results:
    # <--- FIX: Added .reset_index() to turn the grouped-by index into columns --->
    all_vol_stats = pd.concat([r["VolStats"].assign(Token=r["Token"]) for r in results]).reset_index()
    all_cross_stats = pd.concat([r["CrossStats"].assign(Token=r["Token"]) for r in results]).reset_index()
    all_flow_stats = pd.concat([r["FlowVolStats"].assign(Token=r["Token"]) for r in results]).reset_index()

    # --- Keep a copy with categorical columns for grouping ---
    all_vol_stats_group = all_vol_stats.copy()
    all_cross_stats_group = all_cross_stats.copy()
    all_flow_stats_group = all_flow_stats.copy()

    # --- Drop 'NA' buckets before averaging ---
    # Now this works because 'Vol_Bucket' is a column
    all_vol_stats_group = all_vol_stats_group[all_vol_stats_group['Vol_Bucket'] != 'NA']
    all_cross_stats_group = all_cross_stats_group[all_cross_stats_group['Vol_Bucket'] != 'NA']
    all_flow_stats_group = all_flow_stats_group[all_flow_stats_group['Vol_Bucket'] != 'NA']

    # --- Keep only numeric columns for mean calculation ---
    all_vol_stats_numeric = all_vol_stats_group.select_dtypes(include=np.number)
    all_cross_stats_numeric = all_cross_stats_group.select_dtypes(include=np.number)
    all_flow_stats_numeric = all_flow_stats_group.select_dtypes(include=np.number)

    # --- Compute mean stats ---
    mean_vol_stats = all_vol_stats_numeric.groupby(all_vol_stats_group['Vol_Bucket']).mean()
    mean_cross_stats = all_cross_stats_numeric.groupby(
        [all_cross_stats_group['Vol_Bucket'], all_cross_stats_group['Transaction Type']]
    ).mean()
    mean_flow_stats = all_flow_stats_numeric.groupby(
        [all_flow_stats_group['Vol_Bucket'], all_flow_stats_group['Flow_Bucket']]
    ).mean()

    # --- Print summaries ---
    print("==== Mean SOL-in-Curve Stats by Volatility ====")
    print(mean_vol_stats)

    print("\n==== Mean Cross Stats (Volatility x Transaction Type) ====")
    print(mean_cross_stats)

    print("\n==== Mean Flow-Volatility Stats ====")
    print(mean_flow_stats)
else:
    print("No tokens had sufficient data to generate statistics.")
```

```python

from numba import jit

# Parameters
ROLL_VOL = 9
ROLL_PRICE = 9
VOL_Z_THRESH = 2.0
PRICE_Z_THRESH = 0.7
STOP_PCT = 0.03
TP_PCT = 0.06
INITIAL_CAPITAL = 10000.0
# --- MODIFIED PARAMETER ---
# SIZE_PCT is replaced with a fixed dollar amount per trade.
SIZE_USD = 3000.0  # Risk $1000 per trade
# --------------------------
FEE_PCT = 0.0005
SLIPPAGE_PCT = 0.0005

@dataclass
class Trade:
    entry_idx: int
    exit_idx: int
    side: str
    entry_price: float
    exit_price: float
    size_tokens: float
    pnl: float
    return_pct: float

@jit(nopython=True)
def _run_backtest_loop(
    prices,
    long_signals,
    short_signals,
    size_usd, # Pass the absolute size as an argument
    initial_capital,
    stop_pct,
    tp_pct,
    fee_pct,
    slippage_pct
):
    """Numba-compiled backtesting loop using fixed dollar position sizing."""
    capital = initial_capital
    equity_curve = np.full_like(prices, np.nan)
    
    trades = np.zeros((len(prices) // 2, 8)) 
    trade_count = 0

    in_position = False
    pos_side = 0
    pos_entry_price = 0.0
    pos_size_tokens = 0.0
    pos_stop_px = 0.0
    pos_tp_px = 0.0
    pos_entry_idx = 0

    for i in range(len(prices) - 1):
        current_price = prices[i]
        next_price = prices[i+1]
        
        if capital <= 0:
            equity_curve[i] = 0
            continue
        
        if in_position:
            # (Exit logic remains the same...)
            exit_trade = False
            exit_price = current_price
            if pos_side == 1:
                if current_price <= pos_stop_px or current_price >= pos_tp_px:
                    exit_trade = True
                    exit_price = max(pos_stop_px, min(current_price, pos_tp_px))
            elif pos_side == -1:
                if current_price >= pos_stop_px or current_price <= pos_tp_px:
                    exit_trade = True
                    exit_price = min(pos_stop_px, max(current_price, pos_tp_px))
            if exit_trade:
                if pos_side == 1:
                    gross_pnl = (exit_price - pos_entry_price) * pos_size_tokens
                else:
                    gross_pnl = (pos_entry_price - exit_price) * pos_size_tokens
                fee = (pos_entry_price * pos_size_tokens + exit_price * pos_size_tokens) * fee_pct
                net_pnl = gross_pnl - fee
                pnl_pct = net_pnl / capital if capital > 0 else 0.0
                capital += net_pnl
                trades[trade_count] = [pos_entry_idx, i, pos_side, pos_entry_price, exit_price, pos_size_tokens, net_pnl, pnl_pct]
                trade_count += 1
                in_position = False
                pos_side = 0

        if not in_position:
            entry_price = 0.0
            usd_to_use = min(size_usd, capital) # Can't use more than we have
            
            if long_signals[i]:
                entry_price = next_price * (1 + slippage_pct)
                if entry_price > 0:
                    # --- MODIFIED SIZING LOGIC ---
                    size_tokens = usd_to_use / entry_price
                    # -----------------------------
                    in_position = True
                    pos_side = 1
                    pos_entry_price = entry_price
                    pos_size_tokens = size_tokens
                    pos_stop_px = entry_price * (1 - stop_pct)
                    pos_tp_px = entry_price * (1 + tp_pct)
                    pos_entry_idx = i

            elif short_signals[i]:
                entry_price = next_price * (1 - slippage_pct)
                if entry_price > 0:
                    # --- MODIFIED SIZING LOGIC ---
                    size_tokens = usd_to_use / entry_price
                    # -----------------------------
                    in_position = True
                    pos_side = -1
                    pos_entry_price = entry_price
                    pos_size_tokens = size_tokens
                    pos_stop_px = entry_price * (1 + stop_pct)
                    pos_tp_px = entry_price * (1 - tp_pct)
                    pos_entry_idx = i
                
        equity_curve[i] = capital

    if len(equity_curve) > 0:
        equity_curve[-1] = capital

    return equity_curve, trades[:trade_count]

def backtest_mean_reversion(df):
    df = df.copy()
    # (Data preparation remains the same...)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values('Date').reset_index(drop=True)
    price = df['Token Price'].astype(float)
    vol = df['Token Amount'].astype(float)
    df['vol_mean'] = vol.rolling(ROLL_VOL, min_periods=5).mean()
    df['vol_std'] = vol.rolling(ROLL_VOL, min_periods=5).std().fillna(0)
    df['vol_z'] = (vol - df['vol_mean']).div(df['vol_std']).replace([np.inf, -np.inf], 0)
    df['price_mean'] = price.rolling(ROLL_PRICE, min_periods=5).mean()
    df['price_std'] = price.rolling(ROLL_PRICE, min_periods=5).std().fillna(0)
    df['price_z'] = (price - df['price_mean']).div(df['price_std']).replace([np.inf, -np.inf], 0)
    ttype = df['Transaction Type'].astype(str).str.lower()
    df['is_buy_tx'] = (ttype == 'buy')
    df['is_sell_tx'] = (ttype == 'sell')
    df['long_signal'] = (df['vol_z'] >= VOL_Z_THRESH) & (df['price_z'] <= -PRICE_Z_THRESH) & df['is_buy_tx']
    df['short_signal'] = (df['vol_z'] >= VOL_Z_THRESH) & (df['price_z'] >= PRICE_Z_THRESH) & df['is_sell_tx']

    # Convert pandas Series to NumPy arrays for Numba
    prices_np = df['Token Price'].to_numpy(dtype=np.float64)
    long_signals_np = df['long_signal'].to_numpy(dtype=np.bool_)
    short_signals_np = df['short_signal'].to_numpy(dtype=np.bool_)
    
    # Run the backtest loop
    equity_curve_np, trades_np = _run_backtest_loop(
        prices_np, long_signals_np, short_signals_np,
        SIZE_USD, INITIAL_CAPITAL, STOP_PCT, TP_PCT, FEE_PCT, SLIPPAGE_PCT
    )

    # Convert results
    equity = pd.Series(equity_curve_np, index=df.index).ffill().dropna()
    trades = []
    for t in trades_np:
        trades.append(Trade(
            entry_idx=int(t[0]), exit_idx=int(t[1]), side='long' if t[2] == 1 else 'short',
            entry_price=t[3], exit_price=t[4], size_tokens=t[5], pnl=t[6], return_pct=t[7]
        ))

    # --- Performance Metrics ---
    final_capital = equity.iloc[-1] if not equity.empty else INITIAL_CAPITAL
    total_trades = len(trades)
    wins = [t for t in trades if t.pnl > 0]
    losses = [t for t in trades if t.pnl <= 0]

    win_rate = len(wins) / total_trades if total_trades > 0 else 0
    total_pnl = sum(t.pnl for t in trades)
    avg_pnl = np.mean([t.pnl for t in trades]) if total_trades > 0 else 0
    gross_profit = sum(t.pnl for t in wins)
    gross_loss = abs(sum(t.pnl for t in losses)) if losses else 0
    profit_factor = gross_profit / gross_loss if gross_loss > 0 else np.inf
    expectancy = (win_rate * (gross_profit/len(wins) if wins else 0)) - \
                 ((1 - win_rate) * (gross_loss/len(losses) if losses else 0))

    returns = equity.pct_change().dropna()
    volatility = returns.std()
    sharpe_ratio = returns.mean() / volatility * np.sqrt(252) if volatility > 0 else 0
    downside_returns = returns[returns < 0]
    downside_vol = downside_returns.std()
    sortino_ratio = returns.mean() / downside_vol * np.sqrt(252) if downside_vol > 0 else 0

    max_drawdown = (equity.cummax() - equity).max() if not equity.empty else 0
    calmar_ratio = (returns.mean()*252) / max_drawdown if max_drawdown > 0 else np.inf

    # Consecutive wins/losses
    pnl_signs = [1 if t.pnl > 0 else -1 for t in trades]
    max_consec_wins, max_consec_losses, cur_wins, cur_losses = 0, 0, 0, 0
    for s in pnl_signs:
        if s == 1:
            cur_wins += 1
            cur_losses = 0
        else:
            cur_losses += 1
            cur_wins = 0
        max_consec_wins = max(max_consec_wins, cur_wins)
        max_consec_losses = max(max_consec_losses, cur_losses)

    perf = {
        'initial_capital': INITIAL_CAPITAL,
        'final_capital': final_capital,
        'total_trades': total_trades,
        'win_rate': win_rate,
        'total_pnl': total_pnl,
        'avg_pnl': avg_pnl,
        'gross_profit': gross_profit,
        'gross_loss': gross_loss,
        'profit_factor': profit_factor,
        'expectancy': expectancy,
        'sharpe_ratio': sharpe_ratio*np.sqrt(252),
        'sortino_ratio': sortino_ratio*np.sqrt(252),
        'volatility': volatility,
        'max_drawdown': max_drawdown,
        'calmar_ratio': calmar_ratio,
        'max_consec_wins': max_consec_wins,
        'max_consec_losses': max_consec_losses
    }

    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(equity.index, equity.values)
    plt.title('Equity Curve')
    plt.xlabel('Index')
    plt.ylabel('Capital (USD)')
    plt.grid(True)
    plt.show()

    return perf, trades, equity, df
```
