```python
# --- User Activity Segmentation ---

# Calculate the number of unique days each wallet was active
user_activity_counts = df.groupby('Wallet Address')['Date'].nunique().rename('Unique_Active_Days')

# Define a function to segment users
def segment_user(days_active):
    if days_active == 1:
        return 'One-Time User'
    elif 2 <= days_active <= 5:
        return 'Casual User'
    else:
        return 'Power User'

# Apply the segmentation
user_segments = user_activity_counts.apply(segment_user)

print("\nUser Segmentation based on Activity:")
print(user_segments.value_counts())

# Plot the distribution of user segments
user_segments.value_counts().plot(kind='pie', autopct='%1.1f%%',
                                  title='User Segment Distribution',
                                  figsize=(8, 8))
plt.ylabel('')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
# 1. Feature Engineering: Create a summary for each wallet
wallet_summary = df.groupby('Wallet Address').agg(
    total_volume_usd=('TransactionValue_USD', 'sum'),
    transaction_count=('Transaction ID', 'count'),
    avg_transaction_value=('TransactionValue_USD', 'mean'),
    unique_platforms=('Platform', 'nunique')
)

# 2. Scale the data: K-Means is sensitive to feature scales
scaler = StandardScaler()
wallet_summary_scaled = scaler.fit_transform(wallet_summary)

# 3. Apply K-Means clustering
# Let's find 4 distinct clusters
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
wallet_summary['cluster'] = kmeans.fit_predict(wallet_summary_scaled)

# 4. Analyze the clusters
# Group by the new cluster label and find the mean of each feature
cluster_analysis = wallet_summary.groupby('cluster').mean()

print("\nK-Means Cluster Analysis (Cluster Centroids):")
print(cluster_analysis)

# Visualize the clusters (example with two features)
plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=wallet_summary,
    x='total_volume_usd',
    y='transaction_count',
    hue='cluster',
    palette='viridis',
    s=100, alpha=0.8
)
plt.xscale('log')
plt.yscale('log')
plt.title('User Clusters based on Volume and Transaction Count')
plt.show()
```

```python
"""
Crypto mean-reversion trading strategy backtest
File: crypto_mean_reversion_strategy.py

Instructions:
- Provide your dataframe as `df` (pandas.DataFrame) with columns:
  ['Date', 'Token Price', 'Token Amount', 'Transaction Type', 'SOL in Curve', 'Outstanding Shares', 'TransactionValue_USD']
- Ensure df['Date'] is datetime and df is sorted ascending by Date.
- Run the script (it will return a small performance summary and show plots).

Strategy idea (mean-reversion + volume spike):
- When a large buy-volume spike occurs but price has dipped (short-term negative z-score), assume temporary imbalance and go long expecting mean-reversion.
- Symmetric short logic for large sell-volume spikes with short-term positive price z-score.
- Fixed stop-loss and take-profit, simple position sizing.

This script is written to be easy to customize and extend (fees, slippage, different signals, hyperparameter search).
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass


# Parameters 

ROLL_VOL = 20                # rolling window for volume statistics (rows)
ROLL_PRICE = 20              # rolling window for price stats
VOL_Z_THRESH = 3.0           # how many std dev above mean counts as a "spike"
PRICE_Z_THRESH = 0.7         # price z-score threshold to qualify (opposite sign)
STOP_PCT = 0.03              # stop loss as fraction (3%)
TP_PCT = 0.06                # take profit as fraction (6%)
INITIAL_CAPITAL = 10000.0    # starting capital in USD
SIZE_PCT = 0.10              # fraction of capital to risk per trade
FEE_PCT = 0.0005             # exchange fee per trade (0.05%)
SLIPPAGE_PCT = 0.0005        # slippage applied on entry price


@dataclass
class Trade:
    entry_idx: int
    exit_idx: int
    side: str           # 'long' or 'short'
    entry_price: float
    exit_price: float
    size_tokens: float
    pnl: float
    return_pct: float



def backtest_mean_reversion(df):
    df = df.copy()
    # basic checks
    assert 'Date' in df.columns
    assert 'Token Price' in df.columns
    assert 'Token Amount' in df.columns

    # prepare
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values('Date').reset_index(drop=True)
    price = df['Token Price'].astype(float)
    vol = df['Token Amount'].astype(float)

    # features
    df['vol_mean'] = vol.rolling(ROLL_VOL, min_periods=5).mean()
    df['vol_std'] = vol.rolling(ROLL_VOL, min_periods=5).std().replace(0, np.nan)
    df['vol_z'] = (vol - df['vol_mean']) / df['vol_std']

    df['price_mean'] = price.rolling(ROLL_PRICE, min_periods=5).mean()
    df['price_std'] = price.rolling(ROLL_PRICE, min_periods=5).std().replace(0, np.nan)
    df['price_z'] = (price - df['price_mean']) / df['price_std']

    # optional liquidity filter using 'SOL in Curve' or 'Outstanding Shares' if present
    if 'SOL in Curve' in df.columns:
        df['sol_curve'] = df['SOL in Curve']
    else:
        df['sol_curve'] = np.nan

    # signals
    # long signal: large buy volume spike AND short-term price is below mean (negative price_z)
    df['is_buy_tx'] = df['Transaction Type'].astype(str).str.lower().eq('buy')
    df['is_sell_tx'] = df['Transaction Type'].astype(str).str.lower().eq('sell')

    df['long_signal'] = (df['vol_z'] >= VOL_Z_THRESH) & (df['price_z'] <= -PRICE_Z_THRESH) & df['is_buy_tx']
    df['short_signal'] = (df['vol_z'] >= VOL_Z_THRESH) & (df['price_z'] >= PRICE_Z_THRESH) & df['is_sell_tx']

    # backtest loop
    capital = INITIAL_CAPITAL
    equity_curve = []
    trades = []
    in_position = False
    position = None  # store dict with side, entry_price, size_tokens, stop_px, tp_px, entry_idx

    for i in range(len(df)-1):
        row = df.iloc[i]
        next_row = df.iloc[i+1]
        current_price = float(row['Token Price'])
        next_price = float(next_row['Token Price'])

        # if not in position, check for entry
        if not in_position:
            if row['long_signal']:
                # determine entry price using next tick price + slippage
                entry_price = next_price * (1 + SLIPPAGE_PCT)
                # size in USD
                usd_to_use = capital * SIZE_PCT
                size_tokens = usd_to_use / entry_price
                stop_px = entry_price * (1 - STOP_PCT)
                tp_px = entry_price * (1 + TP_PCT)
                in_position = True
                position = dict(side='long', entry_price=entry_price, size_tokens=size_tokens, stop_px=stop_px, tp_px=tp_px, entry_idx=i)
            elif row['short_signal']:
                entry_price = next_price * (1 - SLIPPAGE_PCT)
                usd_to_use = capital * SIZE_PCT
                size_tokens = usd_to_use / entry_price
                stop_px = entry_price * (1 + STOP_PCT)
                tp_px = entry_price * (1 - TP_PCT)
                in_position = True
                position = dict(side='short', entry_price=entry_price, size_tokens=size_tokens, stop_px=stop_px, tp_px=tp_px, entry_idx=i)

        else:
            # check exit conditions on the current tick price
            side = position['side']
            entry_price = position['entry_price']
            size_tokens = position['size_tokens']

            # price to check: current price
            px = current_price
            exit = False
            exit_price = px

            if side == 'long':
                if px <= position['stop_px']:
                    # stopped out
                    exit = True
                    exit_price = position['stop_px']  # assume stop executes at stop price
                elif px >= position['tp_px']:
                    exit = True
                    exit_price = position['tp_px']
            else:  # short
                if px >= position['stop_px']:
                    exit = True
                    exit_price = position['stop_px']
                elif px <= position['tp_px']:
                    exit = True
                    exit_price = position['tp_px']

            # optional: time-based exit after N bars (not implemented, but easy to add)

            if exit:
                # calculate pnl
                if side == 'long':
                    gross_pnl = (exit_price - entry_price) * size_tokens
                else:
                    gross_pnl = (entry_price - exit_price) * size_tokens

                # fees: entry + exit
                fee = (entry_price * size_tokens + exit_price * size_tokens) * FEE_PCT
                net_pnl = gross_pnl - fee

                pnl_pct = net_pnl / capital
                capital = capital + net_pnl

                trades.append(Trade(entry_idx=position['entry_idx'], exit_idx=i, side=side, entry_price=entry_price, exit_price=exit_price, size_tokens=size_tokens, pnl=net_pnl, return_pct=pnl_pct))

                in_position = False
                position = None

        equity_curve.append(capital)

    equity = pd.Series(equity_curve, index=df.index[:len(equity_curve)])

    # metrics
    total_trades = len(trades)
    wins = [t for t in trades if t.pnl > 0]
    losses = [t for t in trades if t.pnl <= 0]
    win_rate = len(wins) / total_trades if total_trades>0 else np.nan
    total_pnl = sum(t.pnl for t in trades)
    avg_pnl = np.mean([t.pnl for t in trades]) if total_trades>0 else 0
    max_drawdown = (equity.cummax() - equity).max() if not equity.empty else 0

    perf = {
        'initial_capital': INITIAL_CAPITAL,
        'final_capital': capital,
        'total_trades': total_trades,
        'win_rate': win_rate,
        'total_pnl': total_pnl,
        'avg_pnl': avg_pnl,
        'max_drawdown': max_drawdown
    }

    # plots
    plt.figure(figsize=(10,5))
    plt.plot(equity.index, equity.values)
    plt.title('Equity Curve')
    plt.xlabel('Index')
    plt.ylabel('Capital (USD)')
    plt.grid(True)
    plt.show()

    return perf, trades, equity, df

if __name__ == '__main__':
     perf, trades, equity, df2 = backtest_mean_reversion(df)
     print(perf)
```
