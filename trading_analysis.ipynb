```python

# Plot price line chart with trade arrows
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px

# Filter data for the specific token
token_address = 'BvJVXsuYRC5cSaexMQcGHm3oMfVu1dyboR6H6af2pump'
price_data = df[df['Token Address'] == token_address].copy()
trades_data = trades_df[trades_df['token_address'] == token_address].copy()

# Convert timestamps to datetime for both datasets
price_data['Date'] = pd.to_datetime(price_data['Date'])
trades_data['timestamp'] = pd.to_datetime(trades_data['timestamp'])

# Create the main figure
fig = go.Figure()

# Add price line
fig.add_trace(go.Scatter(
    x=price_data['Date'],
    y=price_data['Token Price'],
    mode='lines',
    name='Price',
    line=dict(color='blue', width=2)
))

# Separate buy and sell trades
buy_trades = trades_data[trades_data['trade_type'] == 'BUY']
sell_trades = trades_data[trades_data['trade_type'] == 'SELL']

# Add buy trade arrows (green, pointing up)
if not buy_trades.empty:
    fig.add_trace(go.Scatter(
        x=buy_trades['timestamp'],
        y=buy_trades['price'],
        mode='markers+text',
        name='Buy Trades',
        marker=dict(
            symbol='triangle-up',
            size=12,
            color='green',
            line=dict(width=2, color='darkgreen')
        ),
        text='BUY',
        textposition='top center',
        textfont=dict(color='darkgreen', size=10)
    ))

# Add sell trade arrows (red, pointing down)
if not sell_trades.empty:
    fig.add_trace(go.Scatter(
        x=sell_trades['timestamp'],
        y=sell_trades['price'],
        mode='markers+text',
        name='Sell Trades',
        marker=dict(
            symbol='triangle-down',
            size=12,
            color='red',
            line=dict(width=2, color='darkred')
        ),
        text='SELL',
        textposition='bottom center',
        textfont=dict(color='darkred', size=10)
    ))

# Update layout
fig.update_layout(
    title=f'Price Chart with Trades - {token_address}',
    xaxis_title='Date',
    yaxis_title='Price',
    hovermode='x unified',
    showlegend=True,
    width=1200,
    height=600
)

# Add hover information
fig.update_traces(
    hovertemplate='<b>%{fullData.name}</b><br>' +
                  'Date: %{x}<br>' +
                  'Price: %{y}<br>' +
                  '<extra></extra>'
)

fig.show()
```
